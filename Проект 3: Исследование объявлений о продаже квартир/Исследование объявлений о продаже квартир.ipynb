## 3. Предобработка данных

```python
data.info()
```

Оценен объём данных в таблице data: 22 колонки и 23 699 записей. Заметно, что в большинстве колонок есть пропуски значений. Пропуски присутствуют как в колонках, куда пользователи сами вписывают данные (это человеческий фактор), так и в колонках, так и в колонках, по которым данные рассчитываются автоматически на основе картографических данных (на эти пропуски важно обратить внимание технической службы, похоже, что сисетма дает сбой при опредлении картографических данных). 

```python
#в колонке is_apartment не везде было указано, что это апартаменты. Заменим пропуски на False методом fillna(): 
data['is_apartment']=data['is_apartment'].fillna(False)

#пропуски в locality_name заменим на слово "неизвестно", а также изменим слово "посЁлок" на "поселок", чтобы избежать дублирования:
data['locality_name']=data['locality_name'].fillna('неизвестно')
data['locality_name'] = data['locality_name'].str.replace('посёлок', 'поселок') 

#в колонке balcony заменим пропуски на 0, а затем заменим значения на целочисленный тип:
data['balcony']=data['balcony'].fillna(0)
data['balcony']=data['balcony'].astype(int)
    
# Заменим пропуски в высоте потолков на медианное значение методом `k` ближайших соседей:
data['ceiling_height']=data.groupby(['airports_nearest', 'city_centers_nearest'])['ceiling_height'].apply(lambda x: x.fillna(x.median()))

#Переведем значения в столбце first_day_exposition в привычный формат даты:
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'])

# Удаляем записи, где отсутствует количество этажей в доме:
data.dropna(subset=['floors_total'], inplace=True)

data.info()
```

```python
#Колонки living_area и kitchen_area имеют пропуски. Проверим, связано ли это  с тем, что это студии:
display((data[data['living_area'].isnull() == True]['studio'].value_counts()))
display((data[data['kitchen_area'].isnull() == True]['studio'].value_counts()))
```

Видим, что лишь малая доля пропусков в колонках living_area и kitchen_area представляют собой студии. Оcтавим пропуски, как есть.

## 4. Расчёты и добавление результатов в таблицу

```python
#Посчитаем и добавим в таблицу цену квадратного метра:
data['price_for_m2']=data['last_price'] / data['total_area']

#Посчитаем и добавим в таблицу день недели публикации объявления:
data['week_day']=pd.DatetimeIndex(data['first_day_exposition']).weekday

#Посчитаем и добавим в таблицу месяц публикации объявления:
data['month']=pd.DatetimeIndex(data['first_day_exposition']).month

#Посчитаем и добавим в таблицу год публикации объявления:
data['year']=pd.DatetimeIndex(data['first_day_exposition']).year

#Добавим в таблицу характеристику этажа квартиры (первый, последний или другой):
import numpy as np
def floor_group(floor, floors_total):
    return np.where(floor==floors_total, 'последний', np.where(floor==1, 'первый', 'другой'))
data['floor_group'] = floor_group(data['floor'], data['floors_total'])

#Добавим в таблицу соотношение жилой и общей площади (k_living), а также отношение площади кухни к общей (k_kitchen):
data['k_living'] = data['living_area'] / data['total_area']
data['k_kitchen'] = data['kitchen_area'] / data['total_area']

data.head()
```

На даграммах размаха можно разглядеть т.н. «ящик с усами». «Усы» находятся на расстоянии равном 1,5 межквартильным размахам (сокращённо IQR) от границ ящика.В размах «усов» попадают нормальные значения, а за пределами находятся выбросы, изображённые точками. Напишем функцию filter для подсчёта границ с учетом 1.5 размаха, чтобы отсечь лишнее, а затем проверим рузультат методом describe() и пострением диаграмм размаха.

```python
def filter (data, column):
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    IQR = q3 - q1
    data_filtered = data.loc[(data[column] < q3 + 1.5*IQR) & (data[column] > q1 - 1.5*IQR), column]
    return data_filtered

for i in columns:
    data[i] = filter (data, i)

data[columns].describe()
```
